```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('spotify_songs.csv')
df.info()
```

```python
sns.histplot(data=df , x=df["track_popularity"])
```

```python
def  mesan(x, z):
    
    for i in range(len(df)):
        z = z + df.loc[i , x] 
        
    return(z/len(df))

print(mesan("tempo", 0))
```
