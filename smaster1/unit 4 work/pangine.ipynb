```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np
df = pd.read_csv( "penguins (1).csv",encoding='windows-1252')
df.info()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
df = df.dropna()
import seaborn as sns
from sklearn.metrics import silhouette_score
#sns.scatterplot(data=df, x='age', y='income')
from sklearn.model_selection import train_test_split
train_df, _ = train_test_split(df, test_size=0.3, random_state=42)
train_df = train_df[["flipper_length_mm", "body_mass_g"]]
from sklearn.cluster import KMeans
model = KMeans(n_clusters=9)
model.fit(train_df)

closser = model.cluster_centers_
sns.scatterplot(data= closser)
```

```python
k_values = range(1, 11)
wcss = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(train_df)
    wcss.append(kmeans.inertia_)


wcss_df = pd.DataFrame({'K': k_values, 'WCSS': wcss})

sns.scatterplot(data= wcss_df, y="WCSS",x= 'K')
silhouette_score(train_df, model.labels_)
```
